```python
import pandas as pd
from urllib.parse import urlparse

# Lees het bestand in een DataFrame met een andere tekencodering
df = pd.read_csv('metaClean43Brightspace', delimiter='\t', encoding='ISO-8859-1')  # Pas de delimiter en encoding aan indien nodig

# Bekijk de kolomnamen
print(df.columns)

# Functie om de titel uit de URL te extraheren
def extract_title_from_url(url):
    path = urlparse(url).path
    # Titel is het laatste deel van het pad, gescheiden door '-'
    title = path.split('/')[-1]
    return title.replace('-', ' ').title()

# Pas de functie toe op de eerste kolom
df['Title'] = df.iloc[:, 0].apply(extract_title_from_url)

# Zet de titels om naar kleine letters
df['Title'] = df['Title'].str.lower()

# Optioneel: sla het bijgewerkte DataFrame op in een nieuw bestand
df.to_csv('metaClean43Brightspace_with_titles.csv', index=False, sep='\t', encoding='ISO-8859-1')

print(df.head())
```

```python
import pandas as pd
from urllib.parse import urlparse

# Lees de eerste bestand in een DataFrame
df1 = pd.read_csv('/Users/daan/Desktop/HVA/Database management/SalesData.txt', delimiter='\t', encoding='ISO-8859-1')

# Functie om de titel uit de URL te extraheren
def extract_title_from_url(url):
    path = urlparse(url).path
    # Titel is het laatste deel van het pad, gescheiden door '-'
    title = path.split('/')[-1]
    return title.replace('-', ' ').title()

# Haal de titels op uit de eerste bestand en zet ze om naar kleine letters
df1['title'] = df1['url'].apply(extract_title_from_url).str.lower()

# Lees de tweede bestand in een DataFrame
df2 = pd.read_csv('/Users/daan/Desktop/HVA/Database management/metaClean43Brightspace.txt', delimiter='\t', encoding='ISO-8859-1')

# Haal de titels op uit de tweede bestand en zet ze om naar kleine letters
df2['title'] = df2.iloc[:, 0].apply(extract_title_from_url).str.lower()

# Combineer de titels uit beide bestanden en verwijder dubbele waarden
all_titles = pd.concat([df1['title'], df2['title']]).drop_duplicates().reset_index(drop=True)

# Voeg een Movie_ID toe aan de unieke titels
titles_df = pd.DataFrame(all_titles, columns=['title'])
titles_df['Movie_ID'] = range(1, len(titles_df) + 1)

# Voeg de Movie_ID toe aan de oorspronkelijke DataFrames
df1 = df1.merge(titles_df, on='title', how='left')
df2 = df2.merge(titles_df, on='title', how='left')

# Sla de bijgewerkte DataFrames op in nieuwe bestanden (optioneel)
df1.to_csv('SalesData_with_ids.csv', index=False, sep='\t', encoding='ISO-8859-1')
df2.to_csv('metaClean43Brightspace_with_ids.csv', index=False, sep='\t', encoding='ISO-8859-1')

print(df1.head())
print(df2.head())
print(titles_df.head())
```

```python
import pandas as pd
from urllib.parse import urlparse

# Lees het bestand SalesData_with_ids.csv in een DataFrame
sales_data_df = pd.read_csv('/Users/daan/Desktop/HVA/Database management/SalesData_with_ids.csv', delimiter='\t', encoding='ISO-8859-1')

# Lees het bestand ExpertReviewsClean43LIWC in een DataFrame
reviews_df = pd.read_csv('/Users/daan/Desktop/HVA/Database management/ExpertReviewsClean43LIWC_with_titles.csv', delimiter='\t', encoding='ISO-8859-1')  # Pas de delimiter en encoding aan indien nodig

# Functie om de titel uit de URL te extraheren
def extract_title_from_url(url):
    path = urlparse(url).path
    # Titel is het laatste deel van het pad, gescheiden door '-'
    title = path.split('/')[-1]
    return title.replace('-', ' ').lower()

# Pas de functie toe op de eerste kolom van de DataFrame
reviews_df['title'] = reviews_df.iloc[:, 0].apply(extract_title_from_url)

# Zorg ervoor dat de kolommen correct zijn
print(reviews_df.columns)
print(sales_data_df.columns)

# Merge met SalesData_with_ids.csv om de Movie_ID toe te voegen
merged_df = reviews_df.merge(sales_data_df[['title', 'Movie_ID']], on='title', how='left')

# Optioneel: sla het bijgewerkte DataFrame op in een nieuw bestand
merged_df.to_csv('ExpertReviewsClean43LIWC_with_movie_ids.csv', index=False, sep='\t', encoding='ISO-8859-1')

print(merged_df.head())
```
